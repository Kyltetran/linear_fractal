```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
T1 = np.array([[0.5, 0, 0], [0, 0.5, 0], [0, 0, 1]])
T2 = np.array([[0.5, 0, 0.5], [0, 0.5, 0], [0, 0, 1]])
T3 = np.array([[0.5, 0, 0.25], [0, 0.5, np.sqrt(3)/4], [0, 0, 1]])
transformations = [T1, T2, T3]
```

```python
def apply_transformation(point, transformation):
    return np.dot(transformation, point)

# Start with an initial point
point = np.array([0.5, 0.5, 1])
points = [point[:2]]  # Store only x, y for plotting

iterations = 10000
for _ in range(iterations):
    transformation_index = np.random.choice(len(transformations))
    transformation = transformations[transformation_index]
    point = apply_transformation(point, transformation)
    points.append(point[:2])  # Append only x, y
```

```python
points = np.array(points)
plt.scatter(points[:, 0], points[:, 1], s=0.1)
plt.axis('equal')
plt.show()
```
