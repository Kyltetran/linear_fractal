```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def julia_set(c, n=1024, iterations=600, zoom=1):
    x = np.linspace(-1.5, 1.5, n) * zoom
    y = np.linspace(-1.5, 1.5, n) * zoom
    X, Y = np.meshgrid(x, y)
    Z = X + 1j * Y

    img = np.zeros(Z.shape, dtype=float)
    for i in range(iterations):
        Z = Z**2 + c
        mask = np.abs(Z) < 1000
        img += mask

    return img / img.max()
```

```python
# Define the parameter for the Julia set
c = 0.285 + 0.01j
```

```python
# Generate the Julia set image
img = julia_set(c)
```

```python
# Plotting
plt.figure(figsize=(10, 10))
plt.imshow(img, cmap='seismic', extent=[-1.5, 1.5, -1.5, 1.5])
plt.colorbar()
plt.title(f"Julia Set for c = {c}")
plt.show()
```
